## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a dataframe for the Top Spenders grouped by player name
ts_by_players = purchases_df.groupby(["SN"])
#Find the total purchase count
ts_purchase_count = ts_by_players["Item ID"].count()
#Find the Total Value of purchses by adding the price columns
ts_purchase_totalvalue = ts_by_players["Price"].sum()
#Find the average each player spent by dividing the total they spent by the total purchases
ts_avg_purchase = (ts_purchase_totalvalue / ts_purchase_count)

# Build the summary dataframe 
ts_summary = pd.DataFrame({"Purchase Count": ts_purchase_count,
                         "Average Purchase Price": ts_avg_purchase,
                         "Total Purchase Value": ts_purchase_totalvalue})
#To get it in descending order, you have to write ascending=false 
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html

ts_summary = ts_summary.sort_values("Total Purchase Value", ascending=False)
ts_summary["Average Purchase Price"] = ts_summary["Average Purchase Price"]  
ts_summary["Total Purchase Value"] = ts_summary["Total Purchase Value"] 
ts_summary = ts_summary[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

ts_summary.head() 
#If there's time, come back to this to add the .map("${:.2f}".format to each 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Most Popular Items
#Group by item ID then by Item name
all_items_group = purchases_df.groupby(["Item ID", "Item Name"])
#Calculate the number of times the item was purchased
all_items_count = all_items_group['SN'].count()
#add up the sum of the total amount spent for that item
all_items_sum = all_items_group['Price'].sum()
#find the total amount made/spent on that item divided by the times it was purchased for the item price
item_total_price = (all_items_sum / all_items_count)
#find the total purchase total amount made/spent on that item multiplied by the times it was purchased for the item price
item_total_purchase_value = (item_total_price * all_items_count)

# display a summary data frame 
item_summary = pd.DataFrame({"Purchase Count": all_items_count,
                          "Item Price": item_total_price,
                          "Total Purchase Value": item_total_purchase_value})

item_summary = item_summary.sort_values("Purchase Count", ascending=False) 
item_summary["Item Price"] = item_summary["Item Price"]
item_summary["Total Purchase Value"] = item_summary["Total Purchase Value"]
item_summary = item_summary[["Purchase Count", "Item Price", "Total Purchase Value"]]
item_summary.head()
```
